```python
"""
This program converts J-scheme USDB TBME to neutron-neutron (nn) M-scheme TBME  and print in terms qubit number.
The qubits are arranged as follows:

qubit     n     l    2*j        2*jz
_______________________________________________________
0         0     2     5         -5
1         0     2     5         +5      
2         0     2     5         -3             
3         0     2     5         +3        
4         0     2     5         -1            
5         0     2     5         +1           
6         1     0     1         -1        
7         1     0     1         +1              
8         0     2     3         -3           
9         0     2     3         +3            
10        0     2     3         -1            
11        0     2     3         +1     

For different 2*m it will print different sets of TBMEs.
For sd-shell, allowed 2*m values are: 0, +/-2, +/-4, +/-6, +/-8, +/-10

"""

from sympy.physics.quantum.cg import CG
import numpy as np

#18O --> n = 2
n = 2
factor = (18/(16+n))**0.0
t = 1
t1 = 1; t2 = 1  #p --> tz = -1 (-1/2); n --> tz = 1 (1/2)
t3 = 1; t4 = 1  #p --> tz = -1 (-1/2); n --> tz = 1 (1/2)



"""
2  2  2  2    0  1  -2.5250
2  2  2  2    1  0  -1.0480
2  2  2  2    2  1  -1.2170
2  2  2  2    3  0  -0.7250
2  2  2  2    4  1  -0.3660
2  2  2  2    5  0  -3.9290


"""
mm = input("Enter 2*m:")
m = int(mm)
j1 = 5; j2 = 5; j3 = 5; j4 = 5

# <5/2 5/2 | V | 5/2 5/2>    
if j1 == 5 and j2 == 5 and j3 == 5 and j4 == 5:
    V = np.array([-2.5250, -1.0480, -1.2170, -0.7250, -0.3660, -3.9290 ])
    J = np.array([0, 1, 2, 3, 4, 5])
    T = np.array([1, 0, 1, 0, 1, 0])
    C = 2    
    #print("m-scheme tbme for <5/2 5/2 | V | 5/2 5/2>")
    

for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    #rint(m1, m2, m3, m4)
                    #print(mm)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        
                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -5:
                            m11 = 0
                        elif m1 == -3:
                            m11 = 2
                        elif m1 == -1:
                            m11 = 4
                        elif m1 == 1:
                            m11 = 5
                        elif m1 == 3:
                            m11 = 3 
                        elif m1 == 5:
                            m11 = 1
                        
                        #define qubit for m2
                        if m2 == -5:
                            m22 = 0
                        elif m2 == -3:
                            m22 = 2
                        elif m2 == -1:
                            m22 = 4
                        elif m2 == 1:
                            m22 = 5
                        elif m2 == 3:
                            m22 = 3
                        elif m2 == 5:
                            m22 = 1
                        
                        #define qubit for m3
                        if m3 == -5:
                            m33 = 0
                        elif m3 == -3:
                            m33 = 2
                        elif m3 == -1:
                            m33 = 4
                        elif m3 == 1:
                            m33 = 5
                        elif m3 == 3:
                            m33 = 3
                        elif m3 == 5:
                            m33 = 1
                            
                        #define qubit for m4
                        if m4 == -5:
                            m44 = 0
                        elif m4 == -3:
                            m44 = 2
                        elif m4 == -1:
                            m44 = 4
                        elif m4 == 1:
                            m44 = 5
                        elif m4 == 3:
                            m44 = 3
                        elif m4 == 5:
                            m44 = 1    
                        print(m11, m22, m33, m44, float(tbme))

"""
2=d5/2;  3=s1/2

2  2  3  3    1  0  -0.9360
2  2  3  3    0  1  -1.2740

"""

j1 = 5; j2 = 5; j3 = 1; j4 = 1
    
# <5/2 5/2 | V | 1/2 1/2>    
if j1 == 5 and j2 == 5 and j3 == 1 and j4 == 1:
    V = np.array([-0.9360, -1.2740])
    J = np.array([1, 0])
    T = np.array([0, 1])
    C = 2    
    #print("m-scheme tbme for <5/2 5/2 | V | 1/2 1/2>")
    
for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        
                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -5:
                            m11 = 0
                        elif m1 == -3:
                            m11 = 2
                        elif m1 == -1:
                            m11 = 4
                        elif m1 == 1:
                            m11 = 5
                        elif m1 == 3:
                            m11 = 3 
                        elif m1 == 5:
                            m11 = 1
                        
                        #define qubit for m2
                        if m2 == -5:
                            m22 = 0
                        elif m2 == -3:
                            m22 = 2
                        elif m2 == -1:
                            m22 = 4
                        elif m2 == 1:
                            m22 = 5
                        elif m2 == 3:
                            m22 = 3
                        elif m2 == 5:
                            m22 = 1
                        
                        #define qubit for m3
                        if m3 == -1:
                            m33 = 6
                        elif m3 == 1:
                            m33 = 7
                        
                            
                        #define qubit for m4
                        if m4 == -1:
                            m44 = 6
                        elif m4 == 1:
                            m44 = 7  
                        print(m11, m22, m33, m44, float(tbme))      
                        
"""
2=d5/2;  3=s1/2

2  3  2  3    2  0  -0.0820
2  3  2  3    2  1  -1.4700
2  3  2  3    3  0  -3.8330
2  3  2  3    3  1   0.1760

"""

j1 = 5; j2 = 1; j3 = 5; j4 = 1
    
# <5/2 1/2 | V | 5/2 1/2>    
if j1 == 5 and j2 == 1 and j3 == 5 and j4 == 1:
    V = np.array([-0.0820, -1.4700, -3.8330, 0.1760])
    J = np.array([2, 2, 3, 3])
    T = np.array([0, 1, 0, 1])
    C = 1    
    #print("m-scheme tbme for <5/2 1/2 | V | 5/2 1/2>")
    
for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                       
                    
                    #define qubit for m1
                    if m1 == -5:
                        m11 = 0
                    elif m1 == -3:
                        m11 = 2
                    elif m1 == -1:
                        m11 = 4
                    elif m1 == 1:
                        m11 = 5
                    elif m1 == 3:
                        m11 = 3 
                    elif m1 == 5:
                        m11 = 1
                        
                    #define qubit for m2
                    if m2 == -1:
                        m22 = 6
                    elif m2 == 1:
                        m22 = 7
                        
                        
                    #define qubit for m3
                    if m3 == -5:
                        m33 = 0
                    elif m3 == -3:
                        m33 = 2
                    elif m3 == -1:
                        m33 = 4
                    elif m3 == 1:
                        m33 = 5
                    elif m3 == 3:
                        m33 = 3
                    elif m3 == 5:
                        m33 = 1    
                            
                    #define qubit for m4
                    if m4 == -1:
                        m44 = 6
                    elif m4 == 1:
                        m44 = 7  
                    print(m11, m22, m33, m44, float(tbme))

"""
2=d5/2;  3=s1/2

3  3  3  3    0  1  -2.1820
3  3  3  3    1  0  -3.3020

"""

j1 = 1; j2 = 1; j3 = 1; j4 = 1
    
# <1/2 1/2 | V | 1/2 1/2>    
if j1 == 1 and j2 == 1 and j3 == 1 and j4 == 1:
    V = np.array([-2.1820, -3.3020])
    J = np.array([0, 1])
    T = np.array([1, 0])
    C = 2    
    #print("m-scheme tbme for <1/2 1/2 | V | 1/2 1/2>")
    
for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        
                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -1:
                            m11 = 6
                        elif m1 == 1:
                            m11 = 7
                        
                        #define qubit for m2
                        if m2 == -1:
                            m22 = 6
                        elif m2 == 1:
                            m22 = 7
                        
                        
                        #define qubit for m3
                        if m3 == -1:
                            m33 = 6
                        elif m3 == 1:
                            m33 = 7
                        
                            
                        #define qubit for m4
                        if m4 == -1:
                            m44 = 6
                        elif m4 == 1:
                            m44 = 7  
                        print(m11, m22, m33, m44, float(tbme))

"""
 2  2  1  1    0  1  -3.3970
 2  2  1  1    1  0   2.5110
 2  2  1  1    2  1  -0.7300
 2  2  1  1    3  0   0.9620
"""
j1 = 5; j2 = 5; j3 = 3; j4 = 3



# <3/2 3/2 | V | 5/2 5/2>
if j1 == 5 and j2 == 5 and j3 == 3 and j4 == 3:
    V = np.array([-3.3970, 2.5110, -0.7300, 0.9620])
    J = np.array([0, 1, 2, 3])
    T = np.array([1, 0, 1, 0])
    C =  2
    #print("m-scheme tbme for <5/2 5/2 | V | 3/2 3/2>")

for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:

                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()

                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        #print(m1, m2, m3, m4, cg1, cg2, V[i])

                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -5:
                            m11 = 0
                        elif m1 == -3:
                            m11 = 2
                        elif m1 == -1:
                            m11 = 4
                        elif m1 == 1:
                            m11 = 5
                        elif m1 == 3:
                            m11 = 3
                        elif m1 == 5:
                            m11 = 1

                        #define qubit for m2
                        if m2 == -5:
                            m22 = 0
                        elif m2 == -3:
                            m22 = 2
                        elif m2 == -1:
                            m22 = 4
                        elif m2 == 1:
                            m22 = 5
                        elif m2 == 3:
                            m22 = 3
                        elif m2 == 5:
                            m22 = 1

                        #define qubit for m3
                        #if m3 == -5:
                        #    m33 = 7
                        if m3 == -3:
                            m33 = 8
                        elif m3 == -1:
                            m33 = 10
                        elif m3 == 1:
                            m33 = 11
                        elif m3 == 3:
                            m33 = 9
                        #elif m3 == 5:
                        #    m33 = 12

                        #define qubit for m4
                        if m4 == -3:
                            m44 = 8
                        elif m4 == -1:
                            m44 = 10
                        elif m4 == 1:
                            m44 = 11
                        elif m4 == 3:
                            m44 = 9
                        print(m11, m22, m33, m44, float(tbme))


"""
2  1  2  1    1  0  -5.9960
2  1  2  1    2  0  -4.1260
2  1  2  1    3  0  -0.6640
2  1  2  1    4  0  -4.3520
2  1  2  1    1  1  -0.0990
2  1  2  1    2  1  -0.2780
2  1  2  1    3  1   0.0670
2  1  2  1    4  1  -1.9060

"""

j1 = 5; j2 = 3; j3 = 5; j4 = 3
    
# <5/2 3/2 | V | 5/2 3/2>    
if j1 == 5 and j2 == 3 and j3 == 5 and j4 == 3:
    V = np.array([-5.9960, -4.1260, -0.6640, -4.3520, -0.0990, -0.2780, 0.0670, -1.9060 ])
    J = np.array([1, 2, 3, 4, 1, 2, 3, 4 ])
    T = np.array([0, 0, 0, 0, 1, 1, 1, 1])
    C =  1 
    #print("m-scheme tbme for <5/2 3/2 | V | 5/2 3/2>")

for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                       
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        #print(m1, m2, m3, m4, cg1, cg2, V[i])
                        
                    #if m1 < m2 and m3 < m4:
                    #define qubit for m1
                    if m1 == -5:
                        m11 = 0
                    elif m1 == -3:
                        m11 = 2
                    elif m1 == -1:
                        m11 = 4
                    elif m1 == 1:
                        m11 = 5
                    elif m1 == 3:
                        m11 = 3
                    elif m1 == 5:
                        m11 = 1
                        
                    #define qubit for m2
                    if m2 == -3:
                        m22 = 8
                    elif m2 == -1:
                        m22 = 10
                    elif m2 == 1:
                        m22 = 11
                    elif m2 == 3:
                        m22 = 9
                        
                    #define qubit for m3
                    if m3 == -5:
                        m33 = 0
                    if m3 == -3:
                        m33 = 2
                    elif m3 == -1:
                        m33 = 4
                    elif m3 == 1:
                        m33 = 5
                    elif m3 == 3:
                        m33 = 3
                    elif m3 == 5:
                        m33 = 1        
                            
                    #define qubit for m4
                    if m4 == -3:
                        m44 = 8
                    elif m4 == -1:
                        m44 = 10
                    elif m4 == 1:
                        m44 = 11
                    elif m4 == 3:
                        m44 = 9
                    print(m11, m22, m33, m44, float(tbme))                        

"""
1=d3/2

1  1  1  1    0  1  -0.9580
1  1  1  1    1  0  -0.9310
1  1  1  1    2  1  -0.2470
1  1  1  1    3  0  -2.7530


"""

j1 = 3; j2 = 3; j3 = 3; j4 = 3
    
# <3/2 3/2 | V | 3/2 3/2>    
if j1 == 3 and j2 == 3 and j3 == 3 and j4 == 3:
    V = np.array([-0.9580, -0.9310, -0.2470, -2.7530])
    J = np.array([0, 1, 2, 3 ])
    T = np.array([1, 0, 1, 0])
    C = 2 
    #print("m-scheme tbme for <3/2 3/2 | V | 3/2 3/2>")
    
for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                       
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        #print(m1, m2, m3, m4, cg1, cg2, V[i])
                        
                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -3:
                            m11 = 8
                        elif m1 == -1:
                            m11 = 10
                        elif m1 == 1:
                            m11 = 11
                        elif m1 == 3:
                            m11 = 9
                        
                        #define qubit for m2
                        if m2 == -3:
                            m22 = 8
                        elif m2 == -1:
                            m22 = 10
                        elif m2 == 1:
                            m22 = 11
                        elif m2 == 3:
                            m22 = 9
                        
                        #define qubit for m3
                        if m3 == -3:
                            m33 = 8
                        elif m3 == -1:
                            m33 = 10
                        elif m3 == 1:
                            m33 = 11
                        elif m3 == 3:
                            m33 = 9       
                            
                        #define qubit for m4
                        if m4 == -3:
                            m44 = 8
                        elif m4 == -1:
                            m44 = 10
                        elif m4 == 1:
                            m44 = 11
                        elif m4 == 3:
                            m44 = 9
                        print(m11, m22, m33, m44, float(tbme))                        

"""
1=d3/2   3=1s1/2

1  1  3  3    0  1  -0.9240
1  1  3  3    1  0   0.2310

"""

j1 = 1; j2 = 1; j3 = 3; j4 = 3
    
# <3/2 3/2 | V | 1/2 1/2>    
if j1 == 1 and j2 == 1 and j3 == 3 and j4 == 3:
    V = np.array([-0.9240, 0.2310])
    J = np.array([ 0, 1])
    T = np.array([ 1, 0])
    C = 2 
    #print("m-scheme tbme for <3/2 3/2 | V | 1/2 1/2>")
    

for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        #print(m1, m2, m3, m4, cg1, cg2, V[i])
                        
                    if m1 < m2 and m3 < m4:
                        #define qubit for m1
                        if m1 == -1:
                            m11 = 6
                        elif m1 == 1:
                            m11 = 7
                        
                        #define qubit for m2
                        if m2 == -1:
                            m22 = 6
                        elif m2 == 1:
                            m22 = 7
                        
                        
                        #define qubit for m3
                        if m3 == -3:
                            m33 = 8
                        elif m3 == -1:
                            m33 = 10
                        elif m3 == 1:
                            m33 = 11
                        elif m3 == 3:
                            m33 = 9       
                            
                        #define qubit for m4
                        if m4 == -3:
                            m44 = 8
                        elif m4 == -1:
                            m44 = 10
                        elif m4 == 1:
                            m44 = 11
                        elif m4 == 3:
                            m44 = 9
                        print(m11, m22, m33, m44, float(tbme))                        

"""
1=d3/2   3=1s1/2

1  3  1  3    1  0  -3.4090
1  3  1  3    1  1   0.0040
1  3  1  3    2  0  -1.2660
1  3  1  3    2  1  -0.6190

"""

j1 = 1; j2 = 3; j3 = 1; j4 = 3
    
# <3/2 1/2 | V | 3/2 1/2>    
if j1 == 1 and j2 == 3 and j3 == 1 and j4 == 3:
    V = np.array([-3.4090, 0.0040, -1.2660, -0.6190])
    J = np.array([ 1, 1, 2, 2])
    T = np.array([ 0, 1, 0, 1])
    C = 1 
    #print("m-scheme tbme for <3/2 1/2 | V | 3/2 1/2>")

for m1 in range(-j1, j1+2, 2):
    for m2 in range(-j2, j2+2, 2):
        for m3 in range(-j3, j3+2, 2):
            for m4 in range (-j4, j4+2, 2):
                if m1+ m2 == m3 + m4 == m:
                    
                    #rint(m1, m2, m3, m4)
                    tbme = 0
                    for i in range(len(V)):
                        
                        cg1 = CG(j1/2, m1/2, j2/2, m2/2, J[i], m/2)
                        cg1 = cg1.doit()
                        cg2 = CG(j3/2, m3/2, j4/2, m4/2, J[i], m/2)
                        cg2 = cg2.doit()
                    
                        cg11 = CG(t/2, t1/2, t/2, t2/2, T[i], (t1+t2)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg11 = cg11.doit()
                        cg22 = CG(t/2, t3/2, t/2, t4/2, T[i], (t3+t4)/2)  #pn, tz_p = -1/2, tz_n = 1/2
                        cg22 = cg22.doit()
                        tbme = tbme + C*cg1*cg11*cg2*cg22*V[i]*factor
                        #print(m1, m2, m3, m4, cg1, cg2, V[i])
                        
                    
                    #define qubit for m1
                    if m1 == -1:
                        m11 = 6
                    elif m1 == 1:
                        m11 = 7
                        
                    #define qubit for m2
                    if m2 == -3:
                        m22 = 8
                    elif m2 == -1:
                        m22 = 10
                    elif m2 == 1:
                        m22 = 11
                    elif m2 == 3:
                        m22 = 9
                    
                    #define qubit for m3
                    if m3 == -1:
                        m33 = 6
                    elif m3 == 1:
                        m33 = 7     
                            
                    #define qubit for m4
                    if m4 == -3:
                        m44 = 8
                    elif m4 == -1:
                        m44 = 10
                    elif m4 == 1:
                        m44 = 11
                    elif m4 == 3:
                        m44 = 9
                    print(m11, m22, m33, m44, float(tbme))                                            
```
